## 数据清理

在分析数据前,要将数据整理,清楚错误的,多余的数据.

```python
# 读入 `cancer_data.csv`
import pandas as pd
df = pd.read_csv('cancer_data.csv')
# 用 info() 检查哪些列有缺失值
df.info()
```

对于缺失值,可以用该列均值填充

```python
mean = df['texture_mean'].mean()
df['texture_mean'].fillna(mean,inplace = True)
```

对于重复值,可以用`duplicated`函数检查,若有重复值,则返回`True`

取出重复值,可以用`drop_duplicates`函数

```python
df.duplicated()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.duplicated()
```

可以看到`drop_duplicates`删除了第558行

## 重命名列
由于之前修改了数据集，使其仅包括肿瘤特征的均值，因此每个特征末尾好像不需要 "_mean" 。而且，稍后输入分析还要多耗费时间。我们现在想一些要分配给列的新标签。

```python
#去除`_mean`
new_labels = []
for col in df.columns:
    if '_mean' in col:
        new_labels.append(col[:-5])
    else:
        new_labels.append(col)
new_labels
```

```python
df.columns = new_labels

df.head()
```

```python
df.to_csv('cancer_data_edited.csv',index = False)
```
