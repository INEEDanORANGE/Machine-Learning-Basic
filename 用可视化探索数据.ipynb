```python
# 导入并加载数据
import pandas as pd
% matplotlib inline 
df = pd.read_csv('powerplant_data_edited.csv')
df.info()
```

`matplotlib inline`语句能让你在Notebook内部查看数据图

快速查看Dataframe中所有值列的直方图,可以用`hist()`函数 `figsize`参数可以调整图像大小

```python
df.head()
```

```python
#分号影藏不需要的输出
df.hist(figsize = (8,8));
```

`plot`函数,对`kind`参数赋值`hist`也能达到效果

对`kind`参数赋值 `bar`绘制柱状图

对`kind`参数赋值 `pie`绘制饼状图

对`kind`参数赋值 `box`绘制箱线图

赋值x,y名称与 `kind = scatter` 绘制散点图 eg:`df.plot(x='Tempreature',y = 'Net hourly electrical energy output',kind = 'scatter')`

```python
df['Tempreature'].hist();
df['Tempreature'].plot(kind = 'hist');
```

`pd.plotting.scatter_matrix([df_NAME])`pandas 的这一函数可以快速的展示每个变量间的直方图

```python
pd.plotting.scatter_matrix(df,figsize = (16,16));
```

## 可视化工具数据探索测试题

```python
# 绘制温度与电力输出之间的关系图
df.plot(x='Tempreature',y = 'Net hourly electrical energy output',kind = 'scatter');
```

```python
# 绘制湿度分布图
df['Relative Humidity'].hist();
```
